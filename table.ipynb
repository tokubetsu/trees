```python
import os
import json
import pandas as pd
from scipy import stats
```

```python
filenames = os.listdir('stats/')
filenames
```

```python
data = {}
for el in filenames:
    if '.json' in el:
        with open('stats/' + el, encoding='utf-8') as f:
            data[el[:-5]] = json.load(f)
```

```python
n1 = ['width', 'width_no', 'depth', 'size', 'size_part', 'length']
n2 = ['kl', 'mean', 'pearson', 'p_val']
names = ['corpus', ]
for el1 in n1:
    for el2 in n2:
        names.append(el1 + '_' + el2)
names.extend(['max_d', 'max_l'])
```

```python
data_pd = []
for el in data:
    new_el = [el, ]
    for el1 in n1:
        new_el.extend(data[el][el1])
    new_el.extend(data[el]['min_max'])
    data_pd.append(new_el)
```

```python
df = pd.DataFrame(data_pd, columns=names)
```

```python
df
```

```python
df.to_csv('stats/metrics.csv', sep='\t', encoding='utf-8')
```

```python
df['summ'] = df['width_kl'] + df['width_no_kl'] + df['depth_kl'] + df['size_kl'] + df['size_part_kl'] + df['length_kl']
```
